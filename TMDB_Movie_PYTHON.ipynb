```python
# Sort the data in descending order based on revenue and filter the top 500 movies
df1=df1.sort_values('revenue',ascending=False)
```

```python
df2=df1.head(500)
df2
```

```python
# Calculate the measures of central tendency for budget, revenue, and runtime
def mean_median_mode():
    list=['budget','revenue','runtime']
    for i in list:
        print(f"The mean of the {i} is {df2[i].mean()}")
        print(f"The median of the {i} is {df2[i].median()}")
        print(f"The mode of the {i} is {df2[i].mode()[0]}")
        print('*'*40)
mean_median_mode()
```

```python
def outliers():
    list=['budget','revenue','runtime']
    for i in list:
        print(f"The mean of {i} is, {df2[i].mean()}")
        Q1=df2[i].quantile(0.25)
        Q3=df2[i].quantile(0.75)
        print('value of Q1 and Q3',Q1,',',Q3)
        IQR=Q3-Q1
        print('value of IQR',IQR)
        UF = Q3 + (1.5 * IQR) 
        LF = Q1 - (1.5 * IQR)
        print('value of UF and LF',UF,',',LF)
        print("-----------------------------------------------------")
outliers()  
```

```python
# Perform outlier analysis using box plots
plt.figure(figsize=(8, 5))
plt.subplot(1, 3, 1)
plt.boxplot(df2['budget'], vert=False)
plt.title('Budget Box Plot')

plt.subplot(1, 3, 2)
plt.boxplot(df2['revenue'], vert=False)
plt.title('Revenue Box Plot')

plt.subplot(1, 3, 3)
plt.boxplot(df2['runtime'], vert=False)
plt.title('Runtime Box Plot')

plt.show()
```
